### Exercise 10.1

By adapting your Genetic Algorithm code, developed during the Numerical Exercise 9, write a C++ code to solve the TSP with a **Simulated Annealing** (SA) algorithm. Apply your code to the optimization of a path among 

- 30 cities randomly placed on a circumference
- 30 cities randomly placed inside a square

Show your results via:

- a picture of the length of the best path as a function of the iteration of your algorithm
- a picture of the best path

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gen, path, temp = np.loadtxt("output.circle.dat", unpack=True, skiprows=1)

fig, ax1 = plt.subplots(figsize=[15,10])

ax1.set_xlabel('Iteration',fontsize = 17)
ax1.set_ylabel('Path Length', color="red",fontsize = 17)
ax1.plot(gen, path, color="red")
ax1.tick_params(axis='y', labelcolor="red")

ax2 = ax1.twinx()

ax2.set_ylabel('Temp', color="blue",fontsize = 17)
ax2.plot(gen, temp, color="blue")
ax2.tick_params(axis='y', labelcolor="blue")

fig.tight_layout()

ax1.grid(True)
plt.show()
```

```python
path, cityx, cityy, city_id = np.loadtxt("output.circle.cities.dat", unpack=True, skiprows=1)


fig, ax1 = plt.subplots(figsize=[15,15])

plt.scatter(cityx, cityy, marker="")

city = [cityx, cityy]
city = np.transpose(city)

xs = np.array([city[int(path[0])][0], city[int(path[-1])][0]])
ys = np.array([city[int(path[0])][1], city[int(path[-1])][1]])
plt.plot(xs, ys, marker="")

for i in range(city_id.size-1):
    xs = np.array([city[int(path[i])][0], city[int(path[i+1])][0]])
    ys = np.array([city[int(path[i])][1], city[int(path[i+1])][1]])
    plt.plot(xs, ys, marker="")
    
    
for i in range(city_id.size):
    txt = str(int(city_id[i]))
    x = cityx[i]
    y = cityy[i]
    plt.text(x, y, txt, color="white", 
             fontsize = 17, bbox={"boxstyle": "round" ,'facecolor': 'red', 'alpha': 1, 'pad': 0.2}) 
    
plt.title("Circle World", fontsize = 20)
plt.axis("off")
plt.show();
```

```python
gen, path, temp = np.loadtxt("output.square.dat", unpack=True, skiprows=1)

fig, ax1 = plt.subplots(figsize=[15,10])

ax1.set_xlabel('Iteration',fontsize = 17)
ax1.set_ylabel('Path Length', color="red",fontsize = 17)
ax1.plot(gen, path, color="red")
ax1.tick_params(axis='y', labelcolor="red")

ax2 = ax1.twinx()

ax2.set_ylabel('Temp', color="blue",fontsize = 17)
ax2.plot(gen, temp, color="blue")
ax2.tick_params(axis='y', labelcolor="blue")

fig.tight_layout()

ax1.grid(True)
plt.show()
```

Parallelize with MPI libraries your Simulated Annealing code in order to solve the TSP by performing a *Random Search* with **parallel SA searches of the optimal path**:
